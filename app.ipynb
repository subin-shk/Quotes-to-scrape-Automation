```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
def setup():
    driver = webdriver.Chrome()
    driver.get("https://quotes.toscrape.com/")
    return driver

def teardown(driver):
    driver.quit()
```

```python
def get_quotes():
    driver = setup()
    pTag = driver.find_elements(By.XPATH, '//div[@class="quote"]/span[@itemprop="text"]')
    for val in pTag:
        print(val.text)
    teardown(driver)
get_quotes()
```

```python
def get_authors():
  driver=setup()
  authors=driver.find_elements(By.XPATH, '//div[@class="quote"]/span/small[@itemprop="author"]')
  print("No. of authors: ", len(authors))
  for author in authors:
    print(author.text)
  teardown(driver)
get_authors()
```

```python
def get_tags():
  driver=setup()
  tags=driver.find_elements(By.XPATH, '//div[@class="tags"]/a')
  tlen=len(tags)
  print("No. of tags: ", tlen)
  for t in tags:
        print(t.text)
  teardown(driver)
get_tags()
```

```python
def next_btn():
  driver=setup()
  nextBtn = driver.find_element(By.XPATH,"//li[@class='next']/a")
  nextBtn.click()
  # teardown(driver)
next_btn()
```

```python
def love_btn():
  driver=setup()
  loveBtn =driver.find_element(By.XPATH, '//span[@class="tag-item"]/a[@class="tag" and text()="love"]')
  loveBtn.click()



  try:
    actres=driver.find_element(By.XPATH, "//div[@class='container']/h3")
    actualResult = actres.text
    expectedResult= "Viewing tag: love"
    assert actualResult == expectedResult, 'Failed'
    print("Success!")
  except AssertionError as e:
    print(e)
  
  teardown(driver)
love_btn()

```

```python
def about_des():
  driver=setup()
  about = driver.find_element(By.XPATH, "(//small[@class='author']/following-sibling::a)[1]")
  about.click()

  try:
    actres=driver.find_element(By.XPATH, '(//div[@class="author-details"]//child::strong)[2]')
    actualResult=actres.text
    expectedResult="Description:"
    assert actualResult == expectedResult, "No Description found, failed"
    print("Successfully redirected to about")
  except AssertionError as e:
    print(e)
  teardown(driver)
about_des()
```

```python
def total_quotes():
  driver=setup()
  
  totalQuotes=0
  expectedQuotes=100
  page=1;

  try:
    while (page<11):
        noOfQuotesPerPage=0
        pTag = driver.find_elements(By.XPATH, '//div[@class="quote"]')
        noOfQuotesPerPage=len(pTag)
        totalQuotes=noOfQuotesPerPage+totalQuotes

        if (page==10):
          break
        else:
          nextBtn = driver.find_element(By.XPATH,"//li[@class='next']/a")
          nextBtn.click()

        page += 1

    assert totalQuotes == expectedQuotes, "Failed"
    print("Pass, no of quotes: ", totalQuotes)
  except Exception as e:
    print(f"Error: {e}")
  finally:
   teardown(driver)

total_quotes()
      
     
```

```python
def login():
  driver=setup()
  loginBtn=driver.find_element(By.XPATH, "//p/a[text()='Login']")
  loginBtn.click()
  unameField=driver.find_element(By.XPATH, "//input[@id='username']")
  passField=driver.find_element(By.XPATH, "//input[@id='password']")
  unameField.send_keys("Alice")
  passField.send_keys("password")
  submitBtn=driver.find_element(By.XPATH, "//input[@type='submit']")
  submitBtn.click()
  logout=driver.find_element(By.XPATH, "//p/a[@href='/logout']")
  actualResult=logout.text
  expectedResult="Logout"
  try:
    assert actualResult == expectedResult, "Login unsuccessful"
    print("Login Successful")
  except Exception as e:
    print(e)
  finally:
    teardown(driver)
login()
  
```

```python
def author_birth():
  driver=setup()
  try:
    i=1
    while i<5:
      about = driver.find_element(By.XPATH, f"(//small[@class='author']/following-sibling::a)[{i}]")
      about.click()
      
      title=driver.find_element(By.XPATH,"//h3[@class='author-title']")
      dob=driver.find_element(By.XPATH,"//span[@class='author-born-date']")
      place=driver.find_element(By.XPATH,"//span[@class='author-born-location']")
      print("Name:",title.text)
      print("Date of Birth:",dob.text)
      print("Born",place.text)
      print("-----------------")
      driver.back()
      i += 1
  except Exception as e:
    print(e)
author_birth()
```

```python
def scroll_slowly():
  driver=setup()
  height = driver.execute_script("return document.body.scrollHeight")
  for scroll in range(100,height,100):
      driver.execute_script(f"window.scrollTo(0,{scroll})")
      time.sleep(0.1)
scroll_slowly()
```

```python
def scroll():
  driver=setup()
  # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
  # driver.execute_script("window.scrollBy(0, 500);")
  driver.execute_script("window.scrollBy(0, window.innerHeight);")

scroll()
```

```python
def all():
  driver=setup()
  quotes=driver.find_elements(By.CSS_SELECTOR, "div[class='quote']")
  for i in range(len(quotes)):
    for quote in quotes:
      print(quote.text)
      print("----------------")
  teardown(driver)
all()
```

```python
# same thing but getting elements one by one
def display():
    driver = setup()
    
    quotes = driver.find_elements(By.CLASS_NAME, 'text')  
    authors = driver.find_elements(By.CLASS_NAME, 'author')  
    tags_list = driver.find_elements(By.CLASS_NAME, 'tags')  

    print("-------------------------")

    
    for i in range(len(quotes)):  
        print(f'"{quotes[i].text}"')
        print(f" - by {authors[i].text}")

        tags = tags_list[i].find_elements(By.TAG_NAME, 'a')
        tag_texts = [tag.text for tag in tags]

        if tag_texts:
            print("Tags:", ", ".join(tag_texts))
        else:
            print("Tags: None")

        print("-------------------------")
    
    teardown(driver)

display()
```
